```python
from aenum import MultiValueEnum
f=open('./2/input.txt','r')



class RPS(MultiValueEnum):
    ROCK='A','X',1
    PAPER='B','Y',2
    SCISSORS='C','Z',3

class Result(MultiValueEnum):
    LOSS='X',0
    DRAW='Y',3
    WIN='Z',6
class Hand:
    def chooseWinner(opponentsHand: RPS):
        if(opponentsHand==RPS.ROCK):
            return RPS.PAPER
        elif(opponentsHand==RPS.PAPER):
            return RPS.SCISSORS
        elif(opponentsHand==RPS.SCISSORS):
            return RPS.ROCK
    
    def chooseLoser(opponentsHand: RPS):
        if(opponentsHand==RPS.ROCK):
            return RPS.SCISSORS
        elif(opponentsHand==RPS.PAPER):
            return RPS.ROCK
        elif(opponentsHand==RPS.SCISSORS):
            return RPS.PAPER
    def __init__(self,l: str):
        hand=l.strip('\n').split(' ')
        self.opponentsHand=RPS(hand[0])
        self.myHand=RPS(hand[1])
        self.result=Result(hand[1])

    def evaluate(self):
        result=0
        #LOSS
        if(((self.myHand==RPS.ROCK)and (self.opponentsHand==RPS.PAPER)) or 
        ((self.myHand==RPS.PAPER)and (self.opponentsHand==RPS.SCISSORS)) or 
        ((self.myHand==RPS.SCISSORS)and (self.opponentsHand==RPS.ROCK))):
            result=0
        #WIN
        elif(((self.myHand==RPS.ROCK)and (self.opponentsHand==RPS.SCISSORS)) or 
        ((self.myHand==RPS.PAPER)and (self.opponentsHand==RPS.ROCK)) or 
        ((self.myHand==RPS.SCISSORS)and (self.opponentsHand==RPS.PAPER))):
            result=6
        #DRAW
        else:
            result=3
        return result+self.myHand.values[2]
    def evaluate2(self):
        myHand=0
        if(self.result==Result.WIN):
            myHand=Hand.chooseWinner(self.opponentsHand).values[2]
        elif(self.result==Result.DRAW):
            myHand=self.opponentsHand.values[2]
        elif(self.result==Result.LOSS):
            myHand=Hand.chooseLoser(self.opponentsHand).values[2]
        return myHand+self.result.values[1]
    def __repr__(self) -> str:
        return '{} vs {} Hand({})'.format(self.opponentsHand,self.myHand,self.result)


hands=[]
for l in f.readlines():
    hands.append(Hand(l))
# for h in hands[:3]:
#     print(h)
#     print(h.evaluate())
#     print(h.evaluate2())
print(sum([x.evaluate() for x in hands]))
print(sum([x.evaluate2() for x in hands]))


```

```python
import numpy as np
from functools import reduce
f = open( './3/input.txt' , 'r' )

class Group:
    def __init__(self,rucksacks) -> None:
        self.rucksacks=rucksacks
        self.inAllThree=reduce(np.intersect1d,[list(r.line) for r in rucksacks])[0]
        
    def getPriority(self):
        if(ord(self.inAllThree)>=96):
            return ord(self.inAllThree)-96
        else:
            return ord(self.inAllThree)-38

    def __repr__(self) -> str:
        return '{} --> {}'.format(','.join([x.line for x in self.rucksacks]),self.inAllThree)
class Rucksack:
    def __init__(self,line: str) -> None:
        cleanLine=line.strip('\n')
        self.line=cleanLine
        self.compartment1=cleanLine[:int(len(cleanLine)/2)]
        self.compartment2=cleanLine[int(len(cleanLine)/2):]
        self.inBoth=np.intersect1d(list(self.compartment1),list(self.compartment2))[0]
    def getPriority(self):
        if(ord(self.inBoth)>=96):
            return ord(self.inBoth)-96
        else:
            return ord(self.inBoth)-38
    def __repr__(self) -> str:
        return 'comp1:{}, comp2:{}, inBoth:{}/{} ({})'.format(self.compartment1,self.compartment2,self.inBoth,self.getPriority(),self.line)

rucksacks=[]
for l in f.readlines():
    rucksacks.append(Rucksack(l))

# for r in rucksacks[:10]:
#     print(r)

print(sum([x.getPriority() for x in rucksacks]))
groups=[]
# print(len(rucksacks))
for gr in np.array_split(rucksacks,len(rucksacks)/3):
    groups.append(Group(gr))

# for g in groups[:10]:
#     print(g)
sum([x.getPriority() for x in groups])
```

```python
import numpy as np
from functools import reduce
f = open( './4/input.txt' , 'r' )

class Pair:

    def __init__(self,line :str) -> None:
        self.line=line.strip('\n')
        sections=self.line.split(',')
        self.ranges=[]
        
        for s in sections:
            sectionIdLimits=s.split('-')
            self.ranges.append(range(int(sectionIdLimits[0]),int(sectionIdLimits[1])+1))
        self.intersect=np.intersect1d(self.ranges[0],self.ranges[1])
    
    def isIntersectCoveringOne(self):
        for r in self.ranges:
            if((len(r)==len(self.intersect))and(r == self.intersect).all()):
                return True
        return False
    
    def isOverlapping(self):
        if(len(self.intersect)>0):
            return True
        return False

    def __repr__(self) -> str:
        return '{} -> {} -> {}/{}'.format(self.line,self.intersect,self.isIntersectCoveringOne(),self.isOverlapping())
        
pairs=[]
for l in f.readlines():
    pairs.append(Pair(l))
# for p in pairs[:10]:
#     print(p)
print(sum([1 for x in pairs if(x.isIntersectCoveringOne())]))
print(sum([1 for x in pairs if(x.isOverlapping())]))
```

```python
import re
import numpy as np
import copy
from functools import reduce
f = open( './5/input.txt' , 'r' )


class Stack():
    def __init__(self,splitted) -> None:
        self.id=int(splitted[0])
        self.crates=[Crate(crate) for crate in splitted[1:] if len(crate)==3]
    
    def getTop(self):
        return self.crates[-1].item

    def __repr__(self) -> str:
        return '{} --> {}'.format(self.id,str(self.crates))

class Crate():
    def __init__(self,item: str) -> None:
        self.item=item[1]

    def __repr__(self) -> str:
        return '[{}]'.format(self.item)

class Rearrangement():
    def __init__(self,line: str) -> None:
        self.line=line.strip('\n')
        m=re.search(r'move (\d+) from (\d+) to (\d+)',self.line)
        self.count=int(m.group(1))
        self.fromLoc=int(m.group(2))
        self.toLoc=int(m.group(3))


    def __repr__(self) -> str:
        return '{} -> {},{},{}'.format(self.line,self.count,self.fromLoc, self.toLoc)

class Supplies():
    def __init__(self,lines) -> None:
        self.lines=[l.strip('\n') for l in lines]
        n=4
        splitted=[[l[i:i+n] for i in range(0, len(l), n)]for l in self.lines]
        splitted=[[e.strip() for e in stack]for stack in np.flip(np.array(splitted).T,1)]
        self.stacks=[Stack(stack) for stack in splitted]
        
    def rearrange(self,r):
        for i in range(r.count):
            self.stacks[r.toLoc-1].crates.append(self.stacks[r.fromLoc-1].crates.pop())
    def rearrangeNew(self,r):
        self.stacks[r.toLoc-1].crates.extend(self.stacks[r.fromLoc-1].crates[-1*r.count:])
        del self.stacks[r.fromLoc-1].crates[-1*r.count:]
    def printTop(self):
        print(''.join([stack.getTop() for stack in self.stacks]))
    def __copy__(self):
        return Supplies(self.lines)
    def __repr__(self) -> str:
        #r='\n'.join(self.lines)
        r=str(self.stacks)
        return r

lines =[]
l=f.readline()

while (not l=='\n'):
    lines.append(l)
    l=f.readline()

supplies=Supplies(lines)
supplies2=copy.copy(supplies)
# print(supplies)
rs=[]
for l in f.readlines():
    rs.append(Rearrangement(l))
# print(rs[0])
for r in rs:
    supplies.rearrange(r)
supplies.printTop()
for r in rs:
     supplies2.rearrangeNew(r)
# print(supplies)


# print(supplies)
supplies2.printTop()

```

```python
import re
import numpy as np

f = open( './7/input.txt' , 'r' )

class Node():

    def __init__(self, item,location,size=0,parent=None):
        self.item = item
        self.location=location
        self.size=size
        self.parent = parent
        self.childs = []

    def addChild(self, child):
        self.childs.append(child)
        return child
    
    def getChild(self,childName):
        return next((c for c in self.childs if c.item==childName),None)

    def getSize(self):
        if(self.size==0):
            #traverse all your childs
            sum=0
            for c in self.childs:
                sum=sum+c.getSize()
            return sum
        else:
            return self.size
    
    def isFolder(self):
        if(self.size==0):
            return True
        return False

    def getFoldersSmallerThan(self,size: int):
        folders=[]
        if((self.isFolder()) and (self.getSize()<=size)):
            folders.append(self)
            #print('appended {}'.format(self))
        for c in self.childs:
            for f in c.getFoldersSmallerThan(size):
                if((f.isFolder()) and (f.getSize()<=size)):
                    folders.append(f)
        return folders
    
    def getFoldersBiggerThan(self,size: int):
        folders=[]
        if((self.isFolder()) and (self.getSize()>=size)):
            folders.append(self)
            #print('appended {}'.format(self))
        for c in self.childs:
            for f in c.getFoldersBiggerThan(size):
                if((f.isFolder()) and (f.getSize()>=size)):
                    folders.append(f)
        return folders



    def __str__(self) -> str:
        return '{} --> {}({}/{})'.format(self.location,self.item,self.isFolder(),self.getSize())

    def __repr__(self) -> str:
        r=str(self)+'\n'
        # for c in self.childs:
        #     r=r+repr(c)
        return r

currentLocation=[]
rootNode=None
currentNode=None
for l in f.readlines():
    l=l.strip()
    #print(l)
    match=re.search('\$ cd (.*)',l)
    if(match!=None):
        dir=match.group(1)
        if(dir=='..'):
            currentLocation.pop()
            currentNode=currentNode.parent
        elif(dir=='/'):
            rootNode=Node(dir,currentLocation.copy())
            currentLocation.append(dir)
            currentNode=rootNode
        else:
            currentLocation.append(dir)
            currentNode=currentNode.getChild(dir)
        continue

    match=re.search('\$ ls',l)
    if(match!=None):
        #appendToCurrentLocation=True
        continue

    match=re.search('(dir|\d+) (.*)',l)
    if(match!=None):
        dirOrSize=match.group(1)
        name=match.group(2)
        if(dirOrSize=='dir'):
            currentNode.addChild(Node(name,currentLocation.copy(),0,currentNode))
        else:
            currentNode.addChild(Node(name,currentLocation.copy(),int(dirOrSize),currentNode))
        continue




#print(rootNode.getSize())
folders=rootNode.getFoldersSmallerThan(100000)
#print(folders)
sizes=[f.getSize() for f in folders]
#print(folders)
print(sum(sizes))
#print((rootNode.getSize()-40000000))
folders=rootNode.getFoldersBiggerThan((rootNode.getSize()-40000000))
folders.sort(key=lambda x: x.getSize(),reverse=False)
print(folders[0].getSize())
```

```python
import numpy as np
from helper import aoc2022
day=8
fileName=aoc2022.init(day)
f = open(fileName, 'r')


patchOfTrees=np.array(list(map(lambda x:list(map(int,list(x.strip()))), f.readlines())))


def scenicScore(row: int, column: int):
    height = patchOfTrees[row, column]
    # print(height)
    scenicScores = [0, 0, 0, 0]
    # totheleft
    for i in range(column-1, -1, -1):
        scenicScores[0] += 1
        if (patchOfTrees[row, i] >= height):
            break

    # totheright
    for i in range(column+1, patchOfTrees.shape[1]):
        scenicScores[1] += 1
        if (patchOfTrees[row, i] >= height):
            break
    # totheTop
    for i in range(row-1, -1, -1):
        scenicScores[2] += 1
        if (patchOfTrees[i, column] >= height):
            break
# tothebottom
    for i in range(row+1, patchOfTrees.shape[0]):
        scenicScores[3] +=1
        if (patchOfTrees[i, column] >= height):
            break
# print(scenicScores)
    return np.prod(scenicScores)


def isVisibleFromEdge(row: int, column: int):
    height = patchOfTrees[row, column]
    
    # AmIEdge?
    if ((row == 0) or column == 0):
        return True
    if ((row == patchOfTrees.shape[0]-1) or (column == patchOfTrees.shape[1]-1)):
        return True

    # from left
    if (max(patchOfTrees[row, 0:column]) < height):
        return True

    # from top
    if (max(patchOfTrees[0:row, column]) < height):
        return True

    # from right
    if (max(patchOfTrees[row, column+1:]) < height):
        return True

    # from bottom
    if (max(patchOfTrees[row+1:, column]) < height):
        return True

    
    return False
# print(patchOfTrees.shape)

# print(patchOfTrees[0,1])


assertions = [[0, 0, 2, True],
              [7, 20, 4, True,],
              [98, 86, 1, True],
              [87, 70, 4, False],]

for row, column, height, visible in assertions:
    assert patchOfTrees[row, column] == height, 'Expected {} but was {}'.format(
        height, patchOfTrees[row, column])
    assert isVisibleFromEdge(row, column) == visible, 'Expected {} but was {}'.format(
        visible, isVisibleFromEdge(row, column))

assertions = [[0, 0, 0],
              [1, 1, 2]]


for row, column, score in assertions:
    calcScore = scenicScore(row, column)
    assert calcScore == score, 'Expected {} but was {}'.format(
        score, calcScore)


# print(isVisibleFromEdge(1,3))
# print(isVisibleFromEdge(97,97))
# print(isVisibleFromEdge(1,0))

visibleTrees = np.empty(patchOfTrees.shape, dtype=bool)
scenicScoreTrees = np.empty(patchOfTrees.shape, dtype=int)
for index, x in np.ndenumerate(patchOfTrees):
    # print(index)
    visibleTrees[index[0], index[1]] = isVisibleFromEdge(index[0], index[1])
    scenicScoreTrees[index[0], index[1]] = scenicScore(index[0], index[1])
# with np.printoptions(threshold=np.inf):
#     print(visibleTrees)
# print(visibleTrees[98,:])
print(np.count_nonzero(visibleTrees == True))
print(np.amax(scenicScoreTrees))
```
